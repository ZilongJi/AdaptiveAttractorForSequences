```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import brainpy.math as bm
import sys
sys.path.append('..')
import TwoD_fun

seed = 111
```

```python
def downsample(center, interval = 100):
    # downsample the bump center with interval
    ans = np.zeros((np.floor(center.shape[0]/interval).astype(int)-1, center.shape[1]))
    for i in range(ans.shape[0]):
        ans[i,:] = np.mean(center[i*interval:(i+1)*interval,:], axis = 0)
    return ans
```

## plot example superdiffusive and diffusive trajectories

```python
np.random.seed(seed)
#simulate super-diffusive trajectory for 10 times
mu = 0. #small mu means close to the travelling boundary
gamma = 2 #noise level
N=10
Centertraces_sd = []

for i in range(N):
    # simulate super-diffusive trajectory
    center_trace = bm.as_numpy(TwoD_fun.get_trace(mu, gamma, duration=500, a=0.2, tau=1, tau_v=20, sigma_u=2))
    #downsampel the trajectory to 1/100
    center_trace = downsample(center_trace, interval=100)
    
    Centertraces_sd.append(center_trace)
    
#simulate Brownian-diffusive trajectory for 10 times

mu = 0.5 #small mu means close to the travelling boundary
gamma = 2 #noise level
Centertraces_bd = []

for i in range(N):
    # simulate super-diffusive trajectory
    center_trace = bm.as_numpy(TwoD_fun.get_trace(mu, gamma, duration=500, a=0.2, tau=1, tau_v=20, sigma_u=2))
    #downsampel the trajectory to 1/100
    center_trace = downsample(center_trace, interval=100)
    
    Centertraces_bd.append(center_trace)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(3, 1.5), dpi=300)

labelsize = 8
ticksize = 8
cms = plt.get_cmap('Set3')

#superdiffusion
for i, trace in enumerate(Centertraces_sd):
    color = cms(i)
    axs[0].plot(trace[::2,0], trace[::2,1], 
                color='black', lw=0.5, linestyle='-', alpha=0.9,
                marker='o', markersize=2, markerfacecolor=color, markeredgecolor='none')
axs[0].set_title('Awake-diffusion', fontsize=labelsize)
axs[0].set_xlabel('Strong adapt.', fontsize=labelsize)

#diffusion
for i, trace in enumerate(Centertraces_bd):
    color = cms(i)
    axs[1].plot(trace[::5,0], trace[::5,1], 
                color='black', lw=0.5, linestyle='-', alpha=0.9,
                marker='o', markersize=2, markerfacecolor=color, markeredgecolor='none')
axs[1].set_title('Sleep-diffusion', fontsize=labelsize)
axs[1].set_xlabel('Weak adapt.', fontsize=labelsize)

#for all ax, set the same x and y limits
for ax in axs:
    ax.set_xlim(-0.3*np.pi, 0.3*np.pi)
    ax.set_ylim(-0.3*np.pi, 0.3*np.pi)
    ax.tick_params(axis='both', which='major', labelsize=ticksize)
    ax.set_aspect('equal')
    #ticks off
    ax.set_xticks([])
    ax.set_yticks([])
    #axis thickness
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(1.0)
     
plt.tight_layout()

plt.savefig('./Fig_model_superdiffusion_diffusion_examples.pdf', dpi=300)
```

# plot the histogram of superdiffusion and diffusion

```python
#get step size from center traces
def get_stepsize(Centertraces):
    stepsize = []
    for i in range(len(Centertraces)):
        trace = Centertraces[i]
        stepsize.append(np.linalg.norm(trace[1:]-trace[:-1], axis=1))
    return np.concatenate(stepsize)
```

```python
#plot levy flight and brownian motion step size histogram
fig, ax = plt.subplots(1, 1, figsize=(3, 1.5), dpi=300)

custom_color1 = '#F18D00'
custom_color2 = '#009FB9'

step=100 #the step amount for calculating the step size


stepsize_sd = get_stepsize(Centertraces_sd)

#set 40 bins from min to max
bins = np.linspace(np.min(stepsize_sd), np.max(stepsize_sd), 40)
#getbinsize
binsize = bins[1] - bins[0]

ax.hist(stepsize_sd, alpha=0.7, bins=bins, color=custom_color1, edgecolor='grey', label='awake-diffu.')

stepsize_bd = get_stepsize(Centertraces_bd)

#set binsize according to the binsize of superdiffusion, with min and max calculated from the current stepsize
bins = np.arange(np.min(stepsize_bd), np.max(stepsize_bd), binsize)

ax.hist(stepsize_bd, alpha=0.7, bins=bins, color=custom_color2, edgecolor='grey',label='sleep-diffu.')

#legend put legend  below x label
ax.legend(fontsize=ticksize-1, frameon=False, loc='center', bbox_to_anchor=(0.7, 0.85))
#change y ticks to scientific notation， move the scientific notation a bit left
#ax3.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

#label
ax.set_xlabel('Step size (cm)', fontsize=labelsize)
ax.set_ylabel('Counts', fontsize=labelsize)

#set upper and right axis invisible
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#set x ticks from meters to centimeters
ax.set_xticks([0.0, 0.05, 0.1])
ax.set_xticklabels([0.0, 5, 10])
#set tick size to ticksize  
ax.tick_params(axis='both', which='major', labelsize=ticksize)
plt.tight_layout()

plt.savefig('./Fig_model_superdiffusion_diffusion_hist.pdf', dpi=300)
```

## plot the diffusion exponent (log-log scale)

```python
def get_mean_var(Centertraces,interval):
    #get the mean and variance of the step size with interval
    delta_all = []
    for center_trace in Centertraces:
        x = center_trace[:,0]
        y = center_trace[:,1]
        delta = np.sqrt((x[interval:]-x[:-interval])**2 + (y[interval:]-y[:-interval])**2)
        delta_all.append(delta)
    delta_all = np.concatenate(delta_all)
    N = len(Centertraces)
    
    return [[np.mean(delta_all),np.std(delta_all)]]

def f_1(x, A, B):
    return A*x + B
```

```python
#plot the phase diagram in the first subplot
fig, ax = plt.subplots(1, 1, figsize=(3, 1.5), dpi=300)
labelsize = 8
ticksize = 8
custom_color1 = '#F18D00'
custom_color2 = '#009FB9'


mean_std_sd = []
interval_list = np.array(range(1,20,1))
for interval in interval_list:
    mean_std_sd += get_mean_var(Centertraces_sd,interval)
mean_std_sd = np.array(mean_std_sd)

x0 = np.log(interval_list)
y0 = np.log(mean_std_sd[:,0])
A1, B1 = optimize.curve_fit(f_1, x0, y0)[0]
x1 = np.log(interval_list) + 0.5
y1 = A1 * x1 + B1
ax.plot(np.exp(x1), np.exp(y1), '--', linewidth=1, color = custom_color1, alpha = 0.5)

ax.errorbar(interval_list,mean_std_sd[:,0],mean_std_sd[:,1],
             label = 'awake-diffu. slope='+str(round(A1,2)), 
             color = custom_color1, capsize=1, alpha = 0.5)   

mean_std_bd = []
for interval in interval_list:
    mean_std_bd += get_mean_var(Centertraces_bd,interval)
mean_std_bd = np.array(mean_std_bd)

x0 = np.log(interval_list)
y0 = np.log(mean_std_bd[:,0])
A1, B1 = optimize.curve_fit(f_1, x0, y0)[0]
x1 = np.log(interval_list) + 0.5
y1 = A1 * x1 + B1
ax.plot(np.exp(x1), np.exp(y1), '--', linewidth=1, color = custom_color2, alpha = 0.5)

ax.errorbar(interval_list,mean_std_bd[:,0],mean_std_bd[:,1],
             label = 'sleep-diffu. slope='+str(round(A1,2)), 
             color = custom_color2, capsize=1, alpha = 0.5)

#log-log scale of ax4 and equal axis
ax.set_xscale('log')
ax.set_yscale('log')
#ax4.set_aspect('equal')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend(fontsize=ticksize-1, frameon=False, loc='center', bbox_to_anchor=(0.35, 0.9))

ax.set_xlabel(r'Interval $(\Delta t)$ (ms)',fontsize = labelsize)
ax.set_ylabel('Step size (cm)',fontsize = labelsize)

#set xx ticks
ax.set_xticks([1, 10])
ax.set_xticklabels(['$10^{1}$', '$10^{2}$']) #times 10 sicne we sample each 10 points
#replace the y ticks from meter to cm
ax.set_yticks([0.01,0.1, 1])
ax.set_yticklabels(['$10^0$', '$10^{1}$', '$10^{2}$'])
#set tick size to ticksize  
ax.tick_params(axis='both', which='major', labelsize=ticksize)

plt.tight_layout()

plt.savefig('./Fig_model_superdiffusion_diffusion_loglog.pdf', dpi=300)
```
